# 03 — Augmentation Preview

Demonstrates all six synthetic deviation augmentations on sample QuickDraw images.

```python
import os, sys
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

sys.path.insert(0, os.path.join('..', 'src'))
from dataset import numpy_bitmap_to_image
from augment import (
    apply_rotation, apply_closure_failure, apply_perseveration,
    apply_spatial_disorganization, apply_size_distortion,
    apply_omission, generate_deviation_sample, DEVIATION_NAMES
)
print('Imports OK')
```

```python
# Load a single clean drawing
npy_path = os.path.join('..', 'data', 'raw', 'quickdraw', 'cat.npy')
if os.path.exists(npy_path):
    arr = np.load(npy_path)
    clean_img = numpy_bitmap_to_image(arr[0])
    print('Loaded cat drawing.')
else:
    # Fallback: white canvas with a black square
    clean_img = Image.new('RGB', (224, 224), (255, 255, 255))
    from PIL import ImageDraw
    draw = ImageDraw.Draw(clean_img)
    draw.rectangle([60, 60, 164, 164], outline=0, width=4)
    print('Using synthetic fallback image (run notebook 01 first to get real data).')
```

```python
deviations = [
    ('Original',              clean_img),
    ('Rotation',              apply_rotation(clean_img)),
    ('Closure Failure',       apply_closure_failure(clean_img)),
    ('Perseveration',         apply_perseveration(clean_img)),
    ('Spatial Disorganiz.',   apply_spatial_disorganization(clean_img)),
    ('Size Distortion',       apply_size_distortion(clean_img)),
    ('Omission',              apply_omission(clean_img)),
]

fig, axes = plt.subplots(1, 7, figsize=(21, 4))
for ax, (title, img) in zip(axes, deviations):
    ax.imshow(img)
    ax.set_title(title, fontsize=9)
    ax.axis('off')
plt.suptitle('Individual Deviation Augmentations', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Demonstrate compound generate_deviation_sample
fig, axes = plt.subplots(2, 5, figsize=(20, 8))
for ax in axes.flatten():
    aug, labels = generate_deviation_sample(clean_img)
    active = [DEVIATION_NAMES[i] for i, v in enumerate(labels) if v]
    title = ', '.join(active) if active else 'clean'
    ax.imshow(aug)
    ax.set_title(title, fontsize=7, wrap=True)
    ax.axis('off')
plt.suptitle('Random Compound Deviations (10 samples)', fontsize=12)
plt.tight_layout()
plt.show()
```
